# Brief

Does going to university in a different country affect your mental health? A Japanese international university surveyed its students in 2018 and published a study the following year that was approved by several ethical and regulatory boards.

The study found that international students have a higher risk of mental health difficulties than the general population, and that social connectedness (belonging to a social group) and acculturative stress (stress associated with joining a new culture) are predictive of depression.


Explore the `students` data using PostgreSQL to find out if you would come to a similar conclusion for international students and see if the length of stay is a contributing factor.

Here is a data description of the columns you may find helpful.

| Field Name    | Description                                      |
| ------------- | ------------------------------------------------ |
| `inter_dom`     | Types of students (international or domestic)   |
| `japanese_cate` | Japanese language proficiency                    |
| `english_cate`  | English language proficiency                     |
| `academic`      | Current academic level (undergraduate or graduate) |
| `age`           | Current age of student                           |
| `stay`          | Current length of stay in years                  |
| `todep`         | Total score of depression (PHQ-9 test)           |
| `tosc`          | Total score of social connectedness (SCS test)   |
| `toas`          | Total score of acculturative stress (ASISS test) |

```python
%%sql
SELECT * 
FROM students;
```

```python
%%sql
SELECT stay,
       COUNT(*) AS count_int,
       ROUND(AVG(todep), 2) AS average_phq,
       ROUND(AVG(tosc), 2) AS average_scs,
       ROUND(AVG(toas), 2) AS average_as
FROM students
WHERE inter_dom = 'Inter'
GROUP BY stay
ORDER BY stay DESC
LIMIT 9;
```

##  Conclusion
 
The severity of depression is typically assessed based on the total score obtained from the PHQ-9 test. Here's a general guideline for interpreting the total scores:

- **0-4:** Minimal symptoms. These scores suggest that the individual may not be experiencing significant depression.
- **5-9:** Mild depression. Symptoms are present and may be causing some impairment in daily functioning.
- **10-14:** Moderate depression. Symptoms are more pronounced and may be significantly affecting daily life and functioning.
- **15-19:** Moderately severe depression. Symptoms are severe and likely interfering greatly with daily activities and functioning.
- **20 and above:** Severe depression. Symptoms are very severe and may be indicative of a major depressive episode requiring urgent intervention.


To analyze whether the length of stay impacts the mental health of international students based on the provided results, we can observe the trends in the average scores for mental health indicators (`average_phq`, `average_scs`, `average_as`) across different lengths of stay.

Here's a breakdown of the data:

- For students with a length of stay of 0, 1, and 2, the average scores for mental health indicators vary.
- Students with a length of stay of 3, 4, and 5 show an increase in the average scores for mental health indicators, particularly in `average_phq` and `average_scs`.
- However, students with a length of stay of 6, 7, and 8 show a decrease or stabilization in the average scores for mental health indicators.

Based on this analysis, there might be some impact of the length of stay on the mental health of international students. Specifically, students with longer stays (3 years or more) tend to show higher average scores for mental health indicators compared to those with shorter stays. However, this analysis is preliminary and further statistical analysis or data interpretation may be needed to draw definitive conclusions.
